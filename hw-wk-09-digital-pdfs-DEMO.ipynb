```python
pdfs = glob.glob("all_homework_fuds/*.pdf")
pdfs
```

```python
# For loop, each time it goes through a different PDF untill there's no more.
# Here I'm printing ONLY the file names

for pdf in pdfs:
    print(pdf)
```

```python
# Take a PDF into a CSV. pdf is a string function. 
# We have this string, this file name, and turn it into csv.
# It's printing Alaska becaues we ran the for loop above and it's the last file. 
# https://www.w3schools.com/python/ref_string_replace.asp
# pdf. this is the name of the variable, the one I created in the for loop above. 
out_file = pdf.replace(".pdf", ".csv")
out_file
```

```python
# Do one pdf. Then I insert this into the for loop
convert_into (pdf, out_file, pages="all", output_format = "csv")
```

```python
# Now we do the for loop. I repeat the individual command of conversion from the first file, to the rest

outfiles = []
for pdf in pdfs: 
    out_file = pdf.replace(".pdf", ".csv")
    convert_into (pdf, out_file, pages="all", output_format = "csv")
    # After doing this, I want to remember what csv. name I gave to the function above
    outfiles.append(out_file)
```

```python
print(outfiles)
```

```python
# For loop to make a list of dataframes.
# We're going through all csv names in outfiles and reading them, and turning them into a dataframe

dfs = []
for outfile in outfiles:
    df = pd.read_csv(outfile)
    # df.copy it's a function. Make a copy of dataframes when you're working with lists and for loops
    # Python gets confused if you just add DF on a list. 
    # Here I'm addaing a dataframe to a list
    dfs.append(df.copy())
    
    
```

```python
dfs
```

```python
# Now I have a list of dataframes. Now I can concat

fuds = pd.concat(dfs)
```

```python
fuds.to_csv("fuds_1.csv", encoding = "UTF-8", index = False)
```
